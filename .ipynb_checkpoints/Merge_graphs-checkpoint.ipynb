```python
import yoda
import rivet
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
from matplotlib.ticker import AutoMinorLocator
import os



#funo para fazer o grfico dos dados experimentais do ATLAS
def DataFromExperiment(file, scatter, min = 0, max = 0):
    '''
    Grab data from the Yoda.Scatter2D named scatter from the file yoda, following
    the method that colaborations (ALICE, ATLAS, CMS) save them.

    (str, str) -> (list(np.array))
    '''
    data = yoda.read(file)[scatter]

    if max == 0:
        max = len(data.xVals()) + 1

    x = np.asarray(data.xVals())[min:max]
    y = np.asarray(data.yVals())[min:max]
    err = np.asarray((data.yMaxs() - data.yMins()) / 2)[min:max]
    errx = np.asarray((data.xMaxs() - data.xMins()) / 2)[min:max]
    return [x, y, err, errx]






plt.style.use('plotconfig.mplstyle')


# Useful documentation:
# YODA: https://yoda.hepforge.org/pydoc/
# MATPLOTLIB: https://matplotlib.org/
# NUMPY: https://numpy.org/doc/

# Don't forget to load necessary packages in SAMPA
# source /cvmfs/alice.cern.ch/etc/login.sh
# eval `alienv printenv VO_ALICE@Rivet::2.7.2-alice2-1`


yodapp = [] 
yodaPbPb = []
obs = []




            
#Plot dos dados experimentais            
ATLAS = DataFromExperiment('Data/ATLAS_0-10_502_raa.yoda', '/REF/ATLAS_2018_I1673184/d19-x01-y01')

plt.errorbar(*ATLAS, color='k', fmt='s', ms=1, elinewidth = 1., capsize = 1, capthick = 1, label='ATLAS data 0-10%')       





R = [0.2,0.3,0.4,0.6,0.8,1.0]



for i in range(len(R)):
    yodapp.append('/home/monalisa/Desktop/Mestrado/RAA/Matplotlib_RAA/yoda_merge/vacuum/pp_poseidon_raa_R'+str(R[i])+'.yoda')
    yodaPbPb.append('/home/monalisa/Desktop/Mestrado/RAA/Matplotlib_RAA/yoda_merge/simple/Pb_poseidon_raa_R'+str(R[i])+'.yoda')
    obs.append('/RAA_ATLAS/JetpT_R'+str(R[i]))
    
    
    histos_pp = [yoda.read(yodapp[i])]
    histos_PbPb = [yoda.read(yodaPbPb[i])]
    
    pp_jet = histos_pp[0][obs[i]]
    PbPb_jet = histos_PbPb[0][obs[i]]      
                    
                    
    pp_evtc = histos_pp[0]['/_EVTCOUNT'].sumW()
    pp_xsec = histos_pp[0]['/_XSEC'].point(0).x()
    pp_jet.scaleW(pp_xsec / pp_evtc)
                    
                    
                    
    PbPb_evtc = histos_PbPb[0]['/_EVTCOUNT'].sumW()
    PbPb_xsec = histos_PbPb[0]['/_XSEC'].point(0).x()
    PbPb_jet.scaleW(PbPb_xsec / PbPb_evtc)

                    
          
    raa = PbPb_jet / pp_jet
    
    
    x = np.asarray(raa.xVals())
    y = np.asarray(raa.yVals())
    yerr = np.asarray((raa.yMaxs() - raa.yMins()) / 2)
    xerr = np.asarray((raa.xMaxs() - raa.xMins()) / 2)
                    
                

    arr = [x, y, yerr, xerr]

    minpt = 0


    if (R[i] == 0.2):        

        for k in range(minpt, arr[0].size):
            xhigh = arr[0][k] + arr[3][k]
            xlow = arr[0][k] - arr[3][k]
            yhigh = arr[1][k] + arr[2][k]
            ylow = arr[1][k] - arr[2][k]
            if (k==0):
                plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='b', label='R = 0.2')

            plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='b')
            
            
        plt.plot(arr[0], arr[1], alpha=0.1, color='b', linestyle='--')
    
    if (R[i] == 0.3):
        for k in range(minpt, arr[0].size):
            xhigh = arr[0][k] + arr[3][k]
            xlow = arr[0][k] - arr[3][k]
            yhigh = arr[1][k] + arr[2][k]
            ylow = arr[1][k] - arr[2][k]
            if (k==0):
                plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='tab:purple', label='R = 0.3')
            
            plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='tab:purple')
            
        plt.plot(arr[0], arr[1], alpha=0.1, color='tab:purple', linestyle='--')
            
    if (R[i] == 0.4):
        for k in range(minpt, arr[0].size):
            xhigh = arr[0][k] + arr[3][k]
            xlow = arr[0][k] - arr[3][k]
            yhigh = arr[1][k] + arr[2][k]
            ylow = arr[1][k] - arr[2][k]
            if (k==0):
                plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='g', label='R = 0.4')
            
            plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='g')
            
        plt.plot(arr[0], arr[1], alpha=0.1, color='g', linestyle='--')
            
    if (R[i] == 0.6):
        for k in range(minpt, arr[0].size):
            xhigh = arr[0][k] + arr[3][k]
            xlow = arr[0][k] - arr[3][k]
            yhigh = arr[1][k] + arr[2][k]
            ylow = arr[1][k] - arr[2][k]
            if (k==0):
                plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='r', label='R = 0.6')

            plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='r')
            
        plt.plot(arr[0], arr[1], alpha=0.1, color='r', linestyle='--')
            
    if (R[i] == 0.8):
        for k in range(minpt, arr[0].size):
            xhigh = arr[0][k] + arr[3][k]
            xlow = arr[0][k] - arr[3][k]
            yhigh = arr[1][k] + arr[2][k]
            ylow = arr[1][k] - arr[2][k]
            if (k==0):
                plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='c', label='R = 0.8')
            plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='c')
            
        plt.plot(arr[0], arr[1], alpha=0.1, color='c', linestyle='--')
            
    if (R[i] == 1.0):
        for k in range(minpt, arr[0].size):
            xhigh = arr[0][k] + arr[3][k]
            xlow = arr[0][k] - arr[3][k]
            yhigh = arr[1][k] + arr[2][k]
            ylow = arr[1][k] - arr[2][k]
            if (k == minpt):
                plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='tab:orange', label='R = 1.0')
            plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='tab:orange')
            
        plt.plot(arr[0], arr[1], alpha=0.1, color='tab:orange', linestyle='--')

            

     
            

    
    

# Define plot
plt.axhline(y=1.0, xmin=0., xmax=1200., alpha=0.5, color='k', linestyle='--')
plt.figure(1)





# plt.errorbar(x, y, yerr, xerr, fmt='o')
plt.yscale('log') 
#plt.xscale('log')
plt.xlabel('$p_{T}$ [GeV]')
plt.ylabel('$R_{AA}$')
plt.title('Glauber+Bjorken JEWEL+PYTHIA 0-10%')
plt.legend(loc=1)
plt.xlim(98., 1000.)
plt.ylim(0.35, 1.4)



# Save
#plt.savefig('./imagens/simple_raa_Rs.pdf')
plt.savefig('./Figures/5.02TeV/simple_raa_Rs_zoom.pdf')
plt.savefig('./Figures/5.02TeV/simple_raa_Rs_zoom.png')


#Mostrar grfico
plt.show()
 
                    
    
```

```python
#RAA vUSP FLOW w/o recoil Jewel2.4 for HP
import yoda
#import rivet
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
from matplotlib.ticker import AutoMinorLocator
import os



#funo para fazer o grfico dos dados experimentais do ATLAS
def DataFromExperiment(file, scatter, min = 0, max = 0):
    '''
    Grab data from the Yoda.Scatter2D named scatter from the file yoda, following
    the method that colaborations (ALICE, ATLAS, CMS) save them.

    (str, str) -> (list(np.array))
    '''
    data = yoda.read(file)[scatter]

    if max == 0:
        max = len(data.xVals()) + 1

    x = np.asarray(data.xVals())[min:max]
    y = np.asarray(data.yVals())[min:max]
    err = np.asarray((data.yMaxs() - data.yMins()) / 2)[min:max]
    errx = np.asarray((data.xMaxs() - data.xMins()) / 2)[min:max]
    return [x, y, err, errx]






plt.style.use('plotconfig.mplstyle')


# Useful documentation:
# YODA: https://yoda.hepforge.org/pydoc/
# MATPLOTLIB: https://matplotlib.org/
# NUMPY: https://numpy.org/doc/

# Don't forget to load necessary packages in SAMPA
# source /cvmfs/alice.cern.ch/etc/login.sh
# eval `alienv printenv VO_ALICE@Rivet::2.7.2-alice2-1`


yodapp = [] 
yodaPbPb = []
obs = []
obspp = []




            
#Plot dos dados experimentais            
ATLAS = DataFromExperiment('Data/ATLAS_0-10_502_raa.yoda', '/REF/ATLAS_2018_I1673184/d19-x01-y01')

plt.errorbar(*ATLAS, color='k', fmt='s', ms=2, elinewidth = 1., capsize = 2, capthick = 1, label='ATLAS data')       





#R = [0.2,0.3,0.4,0.6,0.8,1.0]
R = 0.4
CENTS = ['0-10','10-20','30-40','50-60']
#CENTS = ['0-10']



for i in range(len(CENTS)):
    yodapp.append('/home/monalisa/Desktop/Mestrado/RAA/Matplotlib_RAA/yoda_merge/vacuum/pp_poseidon_raa_R'+str(R)+'.yoda')
    yodaPbPb.append('/home/monalisa/Desktop/Mestrado/RAA/Matplotlib_RAA/yoda_merge/vUSPhydro/flow_Pb_502_'+str(CENTS[i])+'_reoff_raa_R'+str(R)+'.yoda')
    
    
    obspp.append('/RAA_ATLAS/JetpT_R'+str(R))
    obs.append('/USPJWL_JETSPEC/JetpT_R'+str(R))
    
    
    
    
    
  
    histos_pp = [yoda.read(yodapp[i])]
    histos_PbPb = [yoda.read(yodaPbPb[i])]
    
    
    pp_jet = histos_pp[0][obspp[i]]
    PbPb_jet = histos_PbPb[0][obs[i]]
    
                    
    pp_evtc = histos_pp[0]['/_EVTCOUNT'].sumW()
    pp_xsec = histos_pp[0]['/_XSEC'].point(0).x()
    pp_jet.scaleW(pp_xsec / pp_evtc)
                    
                    
                    
    PbPb_evtc = histos_PbPb[0]['/_EVTCOUNT'].sumW()
    PbPb_xsec = histos_PbPb[0]['/_XSEC'].point(0).x()
    PbPb_jet.scaleW(PbPb_xsec / PbPb_evtc)    
                    
          
    raa = PbPb_jet / pp_jet
    
    
    
    x = np.asarray(raa.xVals())
    y = np.asarray(raa.yVals())
    yerr = np.asarray((raa.yMaxs() - raa.yMins()) / 2)
    xerr = np.asarray((raa.xMaxs() - raa.xMins()) / 2)                

    arr = [x, y, yerr, xerr]
    
    
    
    
    minpt = 0


    if (CENTS[i] == '0-10'):        

        for k in range(minpt, arr[0].size):
            xhigh = arr[0][k] + arr[3][k]
            xlow = arr[0][k] - arr[3][k]
            yhigh = arr[1][k] + arr[2][k]
            ylow = arr[1][k] - arr[2][k]
            if (k==0):
                plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='b', label='$0-10\%$')

            plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.2, color='b')
            
            
        plt.plot(arr[0], arr[1], alpha=0.1, color='b', linestyle='--')
    
    if (CENTS[i] == '10-20'):
        for k in range(minpt, arr[0].size):
            xhigh = arr[0][k] + arr[3][k]
            xlow = arr[0][k] - arr[3][k]
            yhigh = arr[1][k] + arr[2][k]
            ylow = arr[1][k] - arr[2][k]
            if (k==0):
                plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.5, color='tab:orange', label='$10-20\%$')
            
            plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.5, color='tab:orange')
            
        plt.plot(arr[0], arr[1], alpha=0.1, color='tab:orange', linestyle='--')
            
    if (CENTS[i] == '30-40'):
        for k in range(minpt, arr[0].size):
            xhigh = arr[0][k] + arr[3][k]
            xlow = arr[0][k] - arr[3][k]
            yhigh = arr[1][k] + arr[2][k]
            ylow = arr[1][k] - arr[2][k]
            if (k==0):
                plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.5, color='g', label='$30-40\%$')
            
            plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.5, color='g')
            
        plt.plot(arr[0], arr[1], alpha=0.1, color='g', linestyle='--')    
     
    if (CENTS[i] == '50-60'):
        for k in range(minpt, arr[0].size):
            xhigh = arr[0][k] + arr[3][k]
            xlow = arr[0][k] - arr[3][k]
            yhigh = arr[1][k] + arr[2][k]
            ylow = arr[1][k] - arr[2][k]
            if (k==0):
                plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.5, color='tab:pink', label='$50-60\%$')

            plt.fill_between([xlow, xhigh], ylow, yhigh, alpha=0.5, color='tab:pink')
            
        plt.plot(arr[0], arr[1], alpha=0.1, color='tab:pink', linestyle='--')
            
  

     
            

    
    

# Define plot
plt.axhline(y=1.0, xmin=0., xmax=1200., alpha=0.5, color='k', linestyle='--')
plt.figure(1)





# plt.errorbar(x, y, yerr, xerr, fmt='o')
#plt.yscale('log') 
#plt.xscale('log')
plt.xlabel('$p_{T}$ [GeV]')
plt.ylabel('$R_{AA}$')
#plt.title('Jewel2.4')
plt.text(50., 0.9, 'Jewel2.4 vUSP-hydro w/o recoils')
plt.text(880., 0.9, 'R = 0.4')
plt.text(50., 0.8, '$\sqrt{s_{NN}}=5.02$ TeV')

plt.legend(loc=4)
plt.xlim(15., 1010.)
#plt.ylim(1E-1, 10.)



# Save
#plt.savefig('./imagens/simple_raa_Rs.pdf')
plt.savefig('./Figures/5.02TeV/flow_reoff_raa_jewel2.4_cents_R0.4.pdf')
plt.savefig('./Figures/5.02TeV/flow_reoff_raa_jewel2.4_cents_R0.4.png')



#Mostrar grfico
plt.show()
```
